```
import pandas as pd
```

```
dataSet = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/time-series-19-covid-combined.csv')
dataIndia = dataSet[dataSet['Country/Region'] == 'India']
dates = dataIndia['Date'].values
confirmed = dataIndia['Confirmed'].values
recovered = dataIndia['Recovered'].values
deaths = dataIndia['Deaths'].values
```

```
with open('/COVID19/Data/dataSet.csv', 'w+') as f:
    f.write('Date\tConfirmed\tRecovered\tDeaths\n')

    for i in range(len(dates)):
        f.write(dates[i])
        f.write('\t'),
        f.write(str(confirmed[i]))
        f.write('\t')
        f.write(str(recovered[i]))
        f.write('\t')
        f.write(str(deaths[i]))
        f.write('\n')
    f.close()
```
